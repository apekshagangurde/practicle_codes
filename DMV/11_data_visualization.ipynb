```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
import seaborn as sns
import re

# Load the dataset
data = pd.read_csv("PNQ_AQI.csv")  # Load AQI data into a DataFrame

# Explore structure and content of the data
print(data.head())           # Display the first few rows
print(data.info())           # Display info about columns and data types
print(data.describe())       # Display basic statistics

# Clean column names
data.columns = data.columns.str.replace(' µg/m3', '')  # Remove units from column names

# Convert 'Date' column to datetime format and sort by date
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
data.sort_values(by=['Date'], inplace=True, ignore_index=True)

# Handling 'BDL' (Below Detection Limit) and 'NA' values
for _, col in enumerate(list(data.columns[1:3])):  # Loop through specific pollutant columns
    data[f'{col} BDL'] = data[f'{col}'].apply(lambda x: 1 if 'BDL' in x else 0)  # Flag BDL values
    data[f'{col}'] = data[f'{col}'].apply(lambda x: '0' if 'NA' in x else re.findall(r'\d+', x)[-1])
    data[f'{col}'] = data[f'{col}'].astype(int)  # Convert to integer

# Define function to detect outliers in the dataset
outlier_features = list(data.columns[1:5])  # Select columns to check for outliers

def detect_outliers(df, n, features):
    outlier_indices = []

    for col in features:
        q1 = np.nanpercentile(df[col], 25)  # Calculate Q1
        q3 = np.nanpercentile(df[col], 75)  # Calculate Q3
        iqr = q3 - q1  # Calculate IQR
        outlier_step = 1.5 * iqr  # Define the outlier range

        outlier_list_col = df[(df[col] < q1 - outlier_step) | (df[col] > q3 + outlier_step)].index
        outlier_indices.extend(outlier_list_col)

    outlier_indices = Counter(outlier_indices)
    multiple_outliers = list(k for k, v in outlier_indices.items() if v > n)
    return multiple_outliers

# Identify and drop outliers
Outliers_to_drop = detect_outliers(data, 1, outlier_features)
data.drop(Outliers_to_drop, axis=0, inplace=True)

# Calculate monthly AQI average
monthly_aqi = data.resample('M', on='Date')['AQI'].mean()

# Plot AQI trends
plt.plot(monthly_aqi.index, monthly_aqi.values, color='b', marker='o', label='AQI (Monthly Average)')
plt.title('Monthly AQI Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Average AQI Value')
plt.grid(True)
plt.show()

# Separate line plots for each pollutant
# SO2 Trend
plt.plot(data['Date'], data['SO2'], label='SO2', color='b')
plt.title('SO2 Trend Over Time')
plt.xlabel('Date')
plt.ylabel('SO2 Level')
plt.grid(True)
plt.show()

# NOx Trend
plt.plot(data['Date'], data['Nox'], label='NOx', color='g')
plt.title('NOx Trend Over Time')
plt.xlabel('Date')
plt.ylabel('NOx Level')
plt.grid(True)
plt.show()

# RSPM Trend
plt.plot(data['Date'], data['RSPM'], label='RSPM', color='r')
plt.title('RSPM Trend Over Time')
plt.xlabel('Date')
plt.ylabel('RSPM Level')
plt.grid(True)
plt.show()

# SPM Trend
plt.plot(data['Date'], data['SPM'], label='SPM', color='y')
plt.title('SPM Trend Over Time')
plt.xlabel('Date')
plt.ylabel('SPM Level')
plt.grid(True)
plt.show()

# Bar plot comparing AQI across dates
plt.bar(data['Date'], data['AQI'], color='skyblue')
plt.title('AQI Comparison Across Dates')
plt.xlabel('Date')
plt.ylabel('AQI Value')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Violin plot for distribution of pollutant categories
sns.violinplot(x='variable', y='value', data=data[['AQI', 'SO2', 'Nox', 'RSPM', 'SPM']].melt(), inner='quart')
plt.title('Distribution of AQI Values for Different Pollutant Categories')
plt.xlabel('Pollutant Categories')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Scatter plots showing AQI vs pollutants
# AQI vs SO2
plt.scatter(data['SO2'], data['AQI'], alpha=0.5, color='blue')
plt.title('AQI vs. SO2 Scatter Plot')
plt.xlabel('SO2 Level')
plt.ylabel('AQI')
plt.grid(True)
plt.show()

# AQI vs NOx
plt.scatter(data['Nox'], data['AQI'], alpha=0.5, color='green')
plt.title('AQI vs. NOx Scatter Plot')
plt.xlabel('NOx Level')
plt.ylabel('AQI')
plt.grid(True)
plt.show()

# AQI vs RSPM as bubble chart
plt.scatter(data['RSPM'], data['AQI'], s=data['AQI'], alpha=0.5, color='red')
plt.title('AQI vs. RSPM Bubble Chart')
plt.xlabel('RSPM Level')
plt.ylabel('AQI')
plt.grid(True)
plt.show()

# AQI vs SPM as bubble chart
plt.scatter(data['SPM'], data['AQI'], s=data['AQI'], alpha=0.5, color='purple')
plt.title('AQI vs. SPM Bubble Chart')
plt.xlabel('SPM Level')
plt.ylabel('AQI')
plt.grid(True)
plt.show()
```
