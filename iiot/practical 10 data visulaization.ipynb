```python

import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset('diamonds')
print(data.head())
# Step 1: Visualize the distribution of diamond prices (this can represent industrial product prices)
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], bins=30, kde=True)
plt.title('Distribution of Diamond Prices')
plt.xlabel('Price')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
# Step 2: Boxplot of price by cut type (similar to comparing industry sectors)
plt.figure(figsize=(10, 6))
sns.boxplot(x='cut', y='price', data=data)
plt.title('Diamond Prices by Cut Quality')
plt.xlabel('Cut Quality')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
plt.figure(figsize=(10, 6))
sns.scatterplot(x='carat', y='price', hue='cut', data=data)
plt.title('Carat vs Price of Diamonds')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the 'diamonds' dataset
data = sns.load_dataset('diamonds')
print("First few rows of the dataset:\n", data.head())

# Step 1: Price Distribution Analysis
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], bins=30, kde=True)
plt.title('Distribution of Diamond Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Step 2: Price by Cut Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='cut', y='price', data=data)
plt.title('Diamond Prices by Cut Quality')
plt.xlabel('Cut Quality')
plt.ylabel('Price')
plt.tight_layout()
plt.show()

# Step 3: Relationship between Carat and Price, grouped by Cut
plt.figure(figsize=(10, 6))
sns.scatterplot(x='carat', y='price', hue='cut', data=data)
plt.title('Relationship between Carat and Price of Diamonds')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.tight_layout()
plt.show()

# Step 4: Average Price by Clarity Category
average_price_by_clarity = data.groupby('clarity')['price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='clarity', y='price', data=average_price_by_clarity)
plt.title('Average Diamond Price by Clarity')
plt.xlabel('Clarity')
plt.ylabel('Average Price')
plt.tight_layout()
plt.show()

# Step 5: Summary Statistics
summary_stats = data.describe()
print("\nSummary Statistics:\n", summary_stats)
```

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset (using Seaborn's built-in 'tips' dataset)
data = sns.load_dataset("tips")

# Displaying first few rows to understand the dataset structure
print(data.head())

# Exploratory Data Analysis (EDA)
# Check for missing values
print("\nMissing values:\n", data.isnull().sum())

# Basic statistics of the dataset
print("\nStatistical summary:\n", data.describe())



# Analyzing total bill distribution by day
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="total_bill", data=data)
plt.title("Total Bill Distribution by Day")
plt.show()

# Data Analysis Techniques
# 1. Clustering using KMeans
# Clustering data based on 'total_bill' and 'tip'
X = data[['total_bill', 'tip']]
kmeans = KMeans(n_clusters=3, random_state=42)
data['cluster'] = kmeans.fit_predict(X)

plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x="total_bill", y="tip", hue="cluster", palette="viridis")
plt.title("KMeans Clustering of Total Bill and Tip")
plt.show()

# 2. Regression Analysis
# Predicting 'tip' based on 'total_bill'
X = data[['total_bill']]
y = data['tip']

# Splitting data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Applying Linear Regression
reg = LinearRegression()
reg.fit(X_train, y_train)

# Making predictions
y_pred = reg.predict(X_test)

# Calculating regression performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"\nLinear Regression MSE: {mse:.2f}")
print(f"Linear Regression R^2 Score: {r2:.2f}")

# Plotting Regression Line
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_test['total_bill'], y=y_test, color="blue", label="Actual")
sns.lineplot(x=X_test['total_bill'], y=y_pred, color="red", label="Predicted")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Linear Regression: Total Bill vs Tip")
plt.legend()
plt.show()
```
